### Data preprocessing

```python
outliers_index = []
for i in continous:
        Q1 = df[i].quantile(0.25)  # 1ì‚¬ë¶„ìœ„ìˆ˜ (25th percentile)
        Q3 = df[i].quantile(0.75)  # 3ì‚¬ë¶„ìœ„ìˆ˜ (75th percentile)
        IQR = Q3 - Q1   # ì‚¬ë¶„ìœ„ ë²”ìœ„ (Interquartile Range)
        ll = Q1 -1.5*IQR  # í•˜í•œê°’ (Lower limit)
        ul = Q3+1.5*IQR  # ìƒí•œê°’ (Upper limit)

        # í•˜í•œê°’ë³´ë‹¤ ìž‘ê±°ë‚˜ ìƒí•œê°’ë³´ë‹¤ í° ê°’ì„ ì°¾ìŒ
        if len(df[(df[i] < ll) | (df[i] > ul)]) > 0:
             outliers_index.append(df[(df[i] < ll) | (df[i] > ul)].index.to_list())  # |: ë…¼ë¦¬ OR ì—°ì‚°ìž
print(*outliers_index)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df = pd.DataFrame(sc.fit_transform(df),columns = df.columns)
```

```python
X = df
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1,9):    
    kmeans = KMeans(n_clusters = i,init = "k-means++",random_state = 0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
print(wcss)
```

```python
plt.plot(range(1,9),wcss,marker = '*')
plt.xticks(range(1,9))
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# ë°ì´í„°í”„ë ˆìž„ê³¼ X ì„¤ì •
# df = ... (ë°ì´í„°í”„ë ˆìž„ ì„¤ì •)
# X = df[['feature1', 'feature2', ...]].values  # í´ëŸ¬ìŠ¤í„°ë§í•  íŠ¹ì§• ì„ íƒ

# ì‹¤ë£¨ì—£ ê³„ìˆ˜ë¥¼ ì €ìž¥í•  ë¦¬ìŠ¤íŠ¸ ì´ˆê¸°í™”
silhouette_scores = []

# í´ëŸ¬ìŠ¤í„° ìˆ˜ 2ë¶€í„° 8ê¹Œì§€ì— ëŒ€í•´ ì‹¤ë£¨ì—£ ê³„ìˆ˜ë¥¼ ê³„ì‚°
for i in range(2, 9):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=0)
    kmeans.fit(X)
    cluster_labels = kmeans.labels_
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# ê²°ê³¼ ì¶œë ¥
print(silhouette_scores)

# ì‹¤ë£¨ì—£ ê³„ìˆ˜ ê·¸ëž˜í”„ ê·¸ë¦¬ê¸°
plt.plot(range(2, 9), silhouette_scores, marker='o')
plt.title('Silhouette Method for Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()

# ìµœì ì˜ í´ëŸ¬ìŠ¤í„° ìˆ˜ ì°¾ê¸°
optimal_k = np.argmax(silhouette_scores) + 2  # range ì‹œìž‘ì´ 2ì´ë¯€ë¡œ ì¸ë±ìŠ¤ì— 2ë¥¼ ë”í•¨
print(f"Optimal number of clusters based on silhouette score: {optimal_k}")
```

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# Assuming you have a DataFrame 'df' containing the relevant columns
# Replace this DataFrame creation with your actual data

# Generate some example data (replace this with your actual data)
data = {
    'Alcohol': X.Alcohol,
    'Malic_Acid': X.Malic_Acid,
    'Ash': X.Ash,
    'Ash_Alcanity': X.Ash_Alcanity,
    'Magnesium': X.Magnesium,
    'Total_Phenols': X.Total_Phenols,
    'Flavanoids': X.Flavanoids,
    'Nonflavanoid_Phenols': X.Nonflavanoid_Phenols,
    'Proanthocyanin': X.Proanthocyanins,
    'Color_Intensity': X.Color_Intensity,
    'Hue': X.Hue,
    'OD280': X.OD280,
    'Proline': X.Proline,
    'category': X.category,
}
df = pd.DataFrame(data)

# Create traces for each category
traces = []
for category, color in zip(df['category'].unique(), ['red', 'green', 'blue']):
    subset = df[df['category'] == category]
    trace = go.Scatter3d(
        x=subset['Alcohol'],
        y=subset['Malic_Acid'],
        z=subset['Ash'],  
        mode='markers',
        marker=dict(
            color=color,
            size=5,
            opacity=0.8
        ),
        name=str(category)  # Convert category to string
    )
    traces.append(trace)

# Create subplots
fig = make_subplots(rows=1, cols=1, specs=[[{'type': 'scene'}]])

# Add traces to the subplot
for trace in traces:
    fig.add_trace(trace, row=1, col=1)

# Update layout
fig.update_layout(
    scene=dict(
        xaxis=dict(title='Alcohol'),
        yaxis=dict(title='Malic Acid'),
        zaxis=dict(title='Ash')
    ),
    title='Clusters in 3D'
)

# Show plot
fig.show()
```
