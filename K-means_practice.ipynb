```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA

wine = load_wine()
X = wine.data
y = wine.target

print(wine.DESCR)
print(X, y)
```

```python
# K-means 모델 생성
kmeans = KMeans(n_clusters=3)

# 모델 학습
kmeans.fit(X)

# 클러스터링 결과 확인
labels = kmeans.labels_
#print(labels)
#print("---------------------")
#print(y)

# 클러스터 중심 확인
centroids = kmeans.cluster_centers_
#print(centroids)

# 시각화, 데이터의 차원을 2차원으로 축소한 후, K-means 클러스터링 결과를 시각화하는 과정
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
#print(X_pca)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis')
# x와 y 좌표는 각각 데이터 포인트의 주성분에 대응하며, 이를 통해 각 데이터 포인트를 2차원 공간상에 표현
#plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=300, c='r')
#plt.scatter(centroids[:, 0], centroids[:, 1], marker='o', s=300, c='r', edgecolors='k', label='Centroids')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('K-means Clustering of Wine Data')
plt.show()
```
